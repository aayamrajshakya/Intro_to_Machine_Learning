```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 20 data points (example dataset)
np.random.seed(42)  # For reproducibility
X = np.random.uniform(2, 12, 20)  # Random values for X between 2 and 12
y = 3 * X + 2 + np.random.normal(0, 2, 20)  # y = 3X + 2 with some noise

# Add a column of ones to X for the intercept term
X_prime = np.column_stack((np.ones(len(X)), X))

# Compute theta using least squares with @ and .I for inverse
theta = (np.matrix(X_prime.T @ X_prime).I @ X_prime.T) @ y

# Extract intercept and slope
np.array(theta).flatten()  # Convert to array for easier use
```

```python
# intercept, slope = np.array(theta).flatten()

beta_0, beta_1= np.array(theta).flatten()
slope, intercept = beta_1, beta_0

# Generate the regression line
X_line = np.linspace(min(X), max(X), 100)
y_line = slope * X_line + intercept

# Plot the data points and regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_line, y_line, color='red', label=f'Regression line: y = {slope:.2f}x + {intercept:.2f}')
plt.title('Least Squares Regression with 20 Data Points')
plt.xlabel('X (Independent Variable)')
plt.ylabel('y (Dependent Variable)')
plt.legend()
plt.grid()
plt.show()
```
